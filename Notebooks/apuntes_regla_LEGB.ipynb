## GLOBAL SCOPE

- Los nombres globales son variables asignadas en el nivel superior del archivo del módulo que las contiene.

- Los nombres globales solo deben declararse si se asignan dentro de una función.

- Se puede hacer referencia a los nombres globales dentro de una función sin necesidad de declararlos..

```python
x = 88 

def func():
    x = 99  

func()
print(x)  # Aquí se puede ver como el print coge la variable global, que es a la que tiene acceso.
```

```python
x = 88 
print("x =", x) 

def func():
	global x # aquí se coge el valor 99 por ser el último en aparecer y por ser definido como global.
x = 99  

func()
print("x =", x)  
```

## LEGB rule

```python
y, z = 1, 2  

def all_global():
  global x # puede acceder a la variable x por estar definida como global a pesar de que este dentro de una función sin return.
  x = y + z 

all_global()
print('x = %d, y = %d, z = %d' % (x, y, z))
```

## NESTED SCOPES


```python

x = 99  # Definimos variable global.

def f1():
  x = 88 # definimos variable local.
  def f2(): # f2 está dentro de f1.
    print("x local = ", x) # dentro de la función x = 88
  f2() 
f1()  
print("x global = ", x) # fuera de la función x = 99
```

## Funciones de fábrica: Clausuras

Una **clausura** o una **función de fábrica**, la primera describe una **técnica de programación funcional** y la segunda denota un **patrón de diseño**.

El objeto función en cuestión almacena los valores en los ámbitos que la contienen, independientemente de si dichos ámbitos aún están presentes en la memoria.

```python
def f1():
  x = 88  # variable local
  def f2(): # invocando f1 invocamos f2
    print("x =", x)  
  return f2  # f1 devuelve f2 pero con la variables local como valor de x.

action = f1()  # aquí se nombra la función con una variable.
action()

```

```python
def maker(n):
	def action(x):
		return x ** n # return devuelve el resultado de dos argumentos de distintas funciones.
	return action

f = maker(2) # n = 2
f(3)

f(4) # x = 4, 4**2  = 16
```
