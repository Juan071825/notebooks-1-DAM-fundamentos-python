Ejercicio 1

En este ejercicio, la función f(a), no cumple ninguna función ya que no tiene ningún return, de este modo el código que hay dentro de la función se quedará metido en una "caja negra" y no será empleado en ningún otro proceso. Por lo tanto al printar f(b), el interprete devolverá el valor None. Cuando printamos b, el valor que vemos en pantalla es el de lavariable global llamada b. 

```python
def f(a):	    	# a is assigned to (references) the passed object
	a = 99 		# Changes local variable a only
b = 88	
f(b)				# a and b both reference same 88 initially
print("b = ", b)	# b is not changed
print(f(b))
```

Ejercicio 2

Al no tener la función un return, si queremos printar changer(x,l) obtendremos el valor None, pero al ser l una lista si simplemente llamamos a la función changer(x,l) y luego printamos la variable l, esta se habrá visto afectada por la función y cambiará su valor de [1, 2] a ['spam', 2].  

```python
def changer(a, b):	    # Arguments assigned references to objects
		a = 2			# Changes local name's value only
		b[0] = 'spam' 	# Changes shared object in place
					
x = 1
l = [1, 2]
changer(x, l)           # Caller: Pass immutable and mutable objects
print("l = ", l)        # ['spam', 2]  # x is unchanged, l is different!
assert x == 1, "x es 1"
print(changer(x, l))
```

```python
def changer(a, b):	    # Arguments assigned references to objects
		a = 2			# Changes local name's value only
		b[0] = 'spam' 	# Changes shared object in place	
		return a, b		
x = 1
l = [1, 2]
changer(x, l)           # Caller: Pass immutable and mutable objects
print("l = ", l)        # ['spam', 2]  # x is unchanged, l is different!
assert x == 1, "x es 1"
print(changer(x, l))
print(x)
```

Ejercicio 3

En este caso al hacer una copia de l, al invocar la función changer(x, l[:]) si luego printamos l obtendremos el valor inicial de l y no el que cambia la función.

```python
l = [1, 2]
def changer(x, l):	# Pass a copy, so our 'l' does not change 
    x = 2
    l[0] = 'spam' 

changer(x, l[:])
print("l = ", l) 
```
