Ejercicio 1

En primer lugar creamos la variable a, la cual apunta al objeto de valor 3, después definimos b = a, por lo tanto b también apunta al objeto de valor 3. A continuación definiremos a = 'spam', la cuestión aquí es, ¿b cuanto va a valer?¿3 o spam?

Aunque cambies el objeto al que apunta a, b seguirá apuntando al objeto de valor 3 al que apuntaba a, solo definiendo de nuevo a = b conseguiremos b = 'spam', en otras palabras que b apunte al objeto que contiene el string 'spam'. 

```python
a = 3
b = a               # b es 3
print("b es ", b)

a = 3
b = a
a = 'spam'          # b es 3
print("b es ", b)

a = 'spam'
b = a
print("b es ", b)
```

Ejercicio 3

En este ejercicio tenemos una lista [2, 3, 4] llamada L, luego crearemos una variable M igual a L escribiendo M = L, importante que sea este sentido porque si es L = M python empezará a leer por la derecha y se encontrará una variable no definida. 

Después cambiaremos el valor del primer número de la lista desde L, para darnos cuenta que el valor de M también cambió ¿A que se debe? En este caso estamos cambiando el valor del objeto, de la lista a la que apuntan estas dos variables, por lo que al cambiar algo en L lo cambiaremos en M y viceversa.

Resumen: L y M apuntan a la misma lista, al ejecutar L[posición] = valor o M[posición] = valor, modificaremos el valor de la lista y al printar cualquiera de las dos variables el valor será el cambiado.

```python
L = [2, 3, 4]	    # A mutable object
M = L			    # Make a reference to the same object
L[0] = 24 		    # An in-place change

print("L = ", L)    # [24, 3, 4]
print("M = ", M)    # [24, 3, 4]
```

```python
L = [2, 3, 4]

M = L

M[2] = 9

L
```

Ejercicio 4

Lo que cambia con el ejercicio anterior es que al definir M = L a L sele añade [:], esto crea otro objeto con la misma lista, una copia, a la que apunta M, entonces al cambiar un valor de la lista llamada L, la lista M no se verá afectada.

Resumen: L apunta a la lista [2, 3, 4] que luego será cambiada a [24, 3, 4] y M apunta a una copia de la lista [2, 3, 4].

```python
L = [2, 3, 4]
M = L[:]		    # Make a copy of L (or list(L), copy.copy(L), etc.)
L[0] = 24

print("L = ", L)    # [24, 3, 4]

print("M = ", M)    # [2, 3, 4]  M is not changed

L[:]
```

Shared Referenced and Equality

Ejercicio 5

En este ejercicio lo que hacemos es definir la lista L y luego crear una variable llamada M que apunte a esa misma lista, por lo tanto al igualarlas con L == M obtenemos True. El operador is se refiere a la identidad de objetos, a que los dos apuntan al mismo objeto.

```python
L = [1, 2, 3]
M = L 			    # M and L reference the same object

L == M 			    # True: Same values => operador igualdad de valores
L is M 			    # True: operador identidad de objetos: compara las referencias (los punteros)
```

Ejercicio 6

Aquí lo que pasa es que estamos creando dos listas distintas, dos objetos distintos pero que tienen los mismos valores. De este modo al hacer L == M el resultado es True, por que este operador compara los valores de la lista, pero el operador is da False porque se refiere a si los dos nombres de variable apuntan al mismo objeto, que en este caso no es así.

Resumen: L apunta a una lista de valores [1, 2, 3] y M apunta a otra lista que resulta que tiene los mismos varoles que la anterior [1, 2, 3]

```python
L = [1, 2, 3]
M = [1, 2, 3]	    # M and L reference different objects

print(
    "L == M",
    L == M 			# True: Same values
)

print(
    "L is M", 	    # False: operador identidad de objetos: compara las referencias (los punteros)
    L is M
)
```

Ejercicio 7

Aquí definimos dos variables, X e Y, con el mismo número entero. Al contrario que las listas, los numeros son objetos inmutables por lo tanto al hacer X = 42 e Y = 42 estamos creando dos nombres de variables que apuntan al mismo objeto. El operador == se refiere al valor y el operador is a si son el mismo objeto, ambos dan true en este código

Resumen: X ---> 42 <--- Y
  nombre ---> objeto <--- nombre

```python
X = 42
Y = 42		    # Should be two different objects

X  == Y         # True
X is Y 		    # True: Same object anyhow: caching at work!!
```
